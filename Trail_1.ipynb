```python
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from IPython.display import display

# Load the CSV file
csv_path = r"/content/drive/MyDrive/66e31d6ee96cd_student_resource_3/student_resource 3/dataset/sample_test.csv" # Update this to your actual CSV file path
data = pd.read_csv(csv_path)

# Print the first few rows to check the URLs
print("First few URLs:")
print(data.head())

# Display the first 20 images
for index, url in enumerate(data.iloc[:20, 1]):  # URLs are now in the second column (1)
    try:
        # Ensure the URL is a string and starts with http or https
        if isinstance(url, str) and url.startswith(('http://', 'https://')):
            response = requests.get(url)
            if response.status_code == 200:
                # Open the image from the URL
                image = Image.open(BytesIO(response.content))

                # Display the image
                plt.figure(figsize=(5, 5))
                plt.imshow(image)
                plt.axis('off')
                plt.title(f"Image {index + 1}")
                display(plt.gcf())  # Display the figure
                plt.close()  # Close the figure to avoid displaying it again
            else:
                print(f"Failed to retrieve image {index + 1}. Status code: {response.status_code}")
        else:
            print(f"Invalid URL for image {index + 1}: {url}")
    except Exception as e:
        print(f"Error retrieving image {index + 1}: {e}")
```

```python
# OCR Models Installation and Usage in Google Colab

# Common imports and setup
!pip install Pillow numpy matplotlib pytesseract easyocr transformers
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Function to display image
def display_image(image_path):
    img = Image.open(image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Upload an image (run this cell and upload an image)
from google.colab import files
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Display the uploaded image
display_image(image_path)

# 1. Tesseract OCR
!sudo apt install tesseract-ocr
import pytesseract
from PIL import Image

print("Tesseract OCR:")
print(pytesseract.image_to_string(Image.open(image_path)))

# 2. EasyOCR
import easyocr

reader = easyocr.Reader(['en'])
result = reader.readtext(image_path)
print("\nEasyOCR:")
for detection in result:
    print(detection[1])

# 3. Hugging Face Transformers (TrOCR)
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image

processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-handwritten")

image = Image.open(image_path).convert("RGB")

pixel_values = processor(image, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)

generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
print("\nHugging Face TrOCR:")
print(f"Detected text: {generated_text}")
```
